```python
import control as cnt
import matplotlib.pyplot as plt
import numpy as np

# 주어진 전달함수
num = 100
den = [1, 5, 6]

sys_open_loop = cnt.TransferFunction(num, den)

# 폐루프 전달함수 계산
sys_closed_loop = cnt.feedback(sys_open_loop)

print("폐루프 전달함수:", sys_closed_loop)

# unit step 응답곡선 계산
t, y = cnt.step_response(sys_closed_loop)

# 응답곡선 그리기
plt.figure()
plt.plot(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Unit Step Response")
plt.grid()

# 주파수 응답 (보드선도) 계산
mag, phase, omega = cnt.bode(sys_closed_loop, dB=True, Plot=False)

# 보드선도 그리기
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 12))
ax1.semilogx(omega, mag)
ax1.set(xlabel="Frequency (rad/s)", ylabel="Magnitude (dB)")
ax1.set_title("Bode Magnitude Plot")
ax1.grid(which="both")

ax2.semilogx(omega, phase)
ax2.set(xlabel="Frequency (rad/s)", ylabel="Phase (deg)")
ax2.set_title("Bode Phase Plot")
ax2.grid(which="both")

plt.show()
```
